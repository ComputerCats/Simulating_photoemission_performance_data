```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
from scipy import interpolate
```

DATA FOR FIGS 8 AND 11

```python
abs_flat = pd.read_csv('abs_flat.txt', sep = '\t').to_numpy()
oned_abs = pd.read_csv('oned_abs.txt', sep = '\t').to_numpy()
TwoD_abs = pd.read_csv('TwoD_abs.txt', sep = '\t').to_numpy()
```

```python
fig, ax = plt.subplots()


ax.plot(abs_flat[:, 0], abs_flat[:, 1], color = 'k', linewidth = 3, label = 'Flat')
ax.plot(oned_abs[:, 0], oned_abs[:, 1], color = 'red', linewidth = 3, label = '1D')
ax.plot(TwoD_abs[:, 0], TwoD_abs[:, 1], color = 'blue', linewidth = 3, label = '2D')

ax.axvline(x = 2.32, label = '$\lambda = 532$ nm')

ax.grid()
ax.legend()
ax.set_xlim(left = 1.8, right = 2.4)
#ax.set_ylim(top = 18, bottom = 0)
ax.set_xlabel('$\hbar\omega$, eV', fontsize=14)
ax.set_ylabel('Abs, %', fontsize=14)

plt.savefig('fig_8')
```

FIG. 8. Absorption spectra of photocathode without SPPs, 1D grating and 2D grating.

```python
TwoD_p = pd.read_csv('TwoD_p.txt', sep = '\t').to_numpy()
flat_p = pd.read_csv('flat_p.txt', sep = '\t').to_numpy()
OneD_p = pd.read_csv('OneD_p.txt', sep = '\t').to_numpy()
```

```python
fig, ax = plt.subplots()

ax.plot(flat_p[:, 0], flat_p[:, 1], linewidth = 3, color = 'k', label = '$p$, Flat')
ax.plot(TwoD_p[:, 0], TwoD_p[:, 1], color = 'blue', linewidth = 3, label = '$p_1$, 2D grating')
ax.plot(OneD_p[:, 0], OneD_p[:, 1], color = 'red', linewidth = 3, label = '$p_1$, 1D grating')

ax.plot(TwoD_p[:, 0], TwoD_p[:, 2], color = 'blue', linewidth = 3, label = '$p_2$, 2D grating', linestyle = '--')
ax.plot(OneD_p[:, 0], OneD_p[:, 2], color = 'red', linewidth = 3, label = '$p_2$, 1D grating', linestyle = '--')
ax.axvline(x = 2.32, label = '$\lambda = 532$ nm')
ax.grid()
ax.legend()
ax.set_xlim(left = 1.81, right = 2.4)
#ax.set_ylim(top = 25, bottom = 0)

ax.set_xlabel('$\hbar\omega$, eV', fontsize=14)
ax.set_ylabel('p, %', fontsize=14)

plt.savefig('fig_11')
```

FIG. 11. Contributions of fraction of the field absorbed in the cathode layer of photocathode without SPPs and different segments of photocathodes with 1D and 2D grating: $p_1$ is the fraction of photons absorbed by the semiconductor on top of the step, $p_2$ is the fraction of photons absorbed by the semiconductor deposited at the bottom of the groove. At the wavelength of 532 nm, most of the field is absorbed on top of the step.
