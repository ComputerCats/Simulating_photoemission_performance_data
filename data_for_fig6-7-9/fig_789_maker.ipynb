```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Flat = pd.read_csv('Flat.txt', sep = '\t').to_numpy()
p_polariztion_1D = pd.read_csv('1D-p-polariztion.txt', sep = '\t').to_numpy()
s_polariztion_1D = pd.read_csv('1D-s-polariztion.txt', sep = '\t').to_numpy()

twod_pi2 = pd.read_csv('twod_pi2.txt', sep = '\t').to_numpy()
twod_pi8 = pd.read_csv('twod_pi8.txt', sep = '\t').to_numpy()
twod_pi4 = pd.read_csv('twod_pi4.txt', sep = '\t').to_numpy()

Emittance_flat = pd.read_csv('Emittance_flat.txt', sep = '\t').to_numpy()
Emittance_twod_pi2 = pd.read_csv('Emittance_twod_pi2.txt', sep = '\t').to_numpy()
Emittance_1D = pd.read_csv('Emittance_1D.txt', sep = '\t').to_numpy()
```

```python
fig, ax = plt.subplots()

ax.plot(Flat[:, 0], Flat[:180, 1], color = 'k', label = 'Flat', linewidth = 3)
ax.plot(twod_pi2[:, 0], twod_pi2[:, 1], color = 'blue', label = '2D, $\\theta = 0$', linewidth = 3)
ax.plot(s_polariztion_1D[:, 0], s_polariztion_1D[:, 1], color = 'red', label = '1D, s-polarization', linewidth = 3, linestyle = '--')
ax.plot(p_polariztion_1D[:, 0], p_polariztion_1D[:, 1], color = 'red', label = '1D, p-polarization', linewidth = 3)

ax.grid()
ax.set_xlabel('$\hbar\omega$, eV', fontsize = 14)
ax.set_ylabel('QE, %', fontsize = 14)

ax.set_xlim(right = 2.4, left = 1.8)
ax.set_ylim(top = 16.5, bottom = 0)
ax.axvline(x = 2.32, label = '$\lambda = 532$ nm')

ax.legend()

plt.savefig('fig_6')
```

FIG. 6. The QE of 1D grating is polarization-dependent: p-polarization is directed parallel to the groove, s-polarization is directed across the groove.

```python
fig, ax = plt.subplots()

ax.plot(Emittance_1D[:, 0], Emittance_1D[:, 1], color = 'red', label = '1D', linewidth = 3)
ax.plot(Emittance_flat[:, 0], Emittance_flat[:, 1], color = 'k', label = 'Flat', linewidth = 3)
ax.plot(Emittance_twod_pi2[:, 0], Emittance_twod_pi2[:, 1], color = 'blue', label = '2D', linewidth = 3)

ax.grid()
ax.set_xlabel('$\hbar\omega$, eV', fontsize = 14)
ax.set_ylabel('Emittance, mrad/mm', fontsize = 14)

ax.set_xlim(right = 2.4, left = 1.8)
ax.axvline(x = 2.32, label = '$\lambda = 532$ nm')

ax.legend()
plt.savefig('fig_9')
```

FIG. 9. Emittance for cathodes: cathode with 1D grating, 2D grating and photocathode without SPPs.

```python
fig, ax = plt.subplots()

ax.plot(twod_pi2[:, 0], twod_pi2[:, 1], color = 'blue', label = '2D, $\\theta = 0$', linewidth = 3)
ax.plot(twod_pi4[:, 0], twod_pi4[:, 1], color = 'red', label = '2D, $\\theta = \pi/4$', linewidth = 3, linestyle = '--')
ax.plot(twod_pi8[:, 0], twod_pi8[:, 1], color = 'k', label = '2D, $\\theta = \pi/8$', linewidth = 3, linestyle = '-.')

ax.grid()
ax.set_xlabel('$\hbar\omega$, eV', fontsize = 14)
ax.set_ylabel('QE, %', fontsize = 14)

ax.set_xlim(right = 2.4, left = 1.8)
ax.set_ylim(top = 16.5, bottom = 0)
ax.axvline(x = 2.32, label = '$\lambda = 532$ nm')

ax.legend()

plt.savefig('fig_7')
```

FIG. 7. The QE of 2D grating with various incident wave polarization. 2D grating is polarization independent. The legend shows the polarization angle  of the incident light. The differences between the spectra are almost indistinguishable: the average difference between the spectra does not exceed 0.01 %.
